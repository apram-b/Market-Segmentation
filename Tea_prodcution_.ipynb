```python
#Clustering 
from sklearn.cluster import KMeans
```

```python
labels = KMeans(n_clusters=2,random_state=100).fit_predict(df)
```

```python
def cluster_among_variables(x,y,labels):
    
    columns = list(x.columns)
    
    r = int(len(columns)/2)
    if r%2==0:
        r = r
    else:
        r = r+1
        
    f,ax=plt.subplots(r,2,figsize=(15,15))
    a=0
    
    for col in range(0,r):
        for j in range(0,2):
            if a != len(columns):
                ax[col][j].scatter(x[y],x[columns[a]],c=labels)
                ax[col][j].set_xlabel(y)
                ax[col][j].set_ylabel(columns[a])
                a = a+1
```

```python
for col in df.columns:
    cluster_among_variables(df,col,labels)
```
